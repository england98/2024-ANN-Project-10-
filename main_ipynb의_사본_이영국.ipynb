```python
# 데이터 셋 불러오기 - 대기정보
# path_air = '/content/drive/MyDrive/Colab Notebooks/Project_SD/DataSet/AirInfo_2017-2023.csv'
path_air = '/content/AirInfo_2017-2023.csv'
df_air = pd.read_csv(path_air)

# 컬럼명 변경
df_air.rename(columns={'측정소코드': 'StationCode', '측정일시': 'Timestamp'}, inplace=True)

# 전국 측정소의 데이터라서 미리 지정한 측정소 Filter
air_StationCode = [111121, 131571, 131117, 131622, 131444]
df_air = df_air[df_air['StationCode'].isin(air_StationCode)]

# df_air.loc[:, 'Timestamp'] = df_air['Timestamp'].astype('object')
# df_air['Timestamp'] = pd.to_datetime(df_air['Timestamp'])
# df_air['Timestamp'] = df_air['Timestamp'].dt.strftime('%Y%m%d%H')
```

```python
# 데이터 셋 불러오기 - 기상정보
path_weather = '/content/drive/MyDrive/Colab Notebooks/Project_SD/DataSet/WeatherInfo_2017_2023.csv'
# df_weather = pd.read_csv(path_weather, encoding='cp949')
df_weather = pd.read_csv(path_weather)

df_weather.rename(columns={
    '지점': 'PointCode',
    '일시': 'Timestamp',
    '기온(°C)': 'Temperature_C',
    '강수량(mm)': 'Precipitation_mm',
    '풍속(m/s)': 'WindSpeed_mps',
    '풍향(16방위)': 'WindDirection_16Points',
    '습도(%)': 'Humidity_p',
    '증기압(hPa)': 'VaporPressure_hPa',
    '이슬점온도(°C)': 'DewPoint_C',
    '현지기압(hPa)': 'LocalPressure_hPa',
    '해면기압(hPa)': 'SeaLevelPressure_hPa',
    '일조(hr)': 'Sunshine_Hours',
    '일사(MJ/m2)': 'SolarRadiation_MJ_m2',
    '전운량(10분위)': 'TotalCloudCoverage_10ths',
    '중하층운량(10분위)': 'MidLowCloudCoverage_10ths',
    '시정(10m)': 'Visibility_10m',
    '지면온도(°C)': 'GroundTemperature_C'
}, inplace=True)

df_weather['Timestamp'] = pd.to_datetime(df_weather['Timestamp'])
df_weather['Timestamp'] = df_weather['Timestamp'].dt.strftime('%Y%m%d%H').astype('int64')
```

```python
# 매핑정보
mapping_data = {
    'Locate': ['종로구', '동두천', '수원시', '양평군', '이천시'],
    'StationCode': [111121, 131571, 131117, 131622, 131444],
    'PointCode': [108, 98, 119, 202, 203]
}

df_mapping = pd.DataFrame(mapping_data)

df_A_filter = pd.merge(df_air, df_mapping[['StationCode', 'PointCode']], on='StationCode')
df_W_filter = pd.merge(df_weather, df_mapping[['StationCode', 'PointCode']], on='PointCode')

df = pd.merge(df_A_filter, df_W_filter, how='inner', on=['StationCode', 'PointCode', 'Timestamp'])
```
